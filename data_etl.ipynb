```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
csv_file = "../Resources/top-500-movies.csv"
money_df = pd.read_csv(csv_file)
money_df.head()
```

```python
csv_file = "../Resources/imdb_top_1000.csv"
info_df = pd.read_csv(csv_file)
info_df
```

### Create new data with select columns

```python
new_money_df = money_df[['rank', 'title','genre', 'production_cost','worldwide_gross','theaters']].copy()
new_money_df.head()
```

```python
new_info_df = info_df[['Series_Title', 'IMDB_Rating','Meta_score','Director','Star1','Star2','No_of_Votes']].copy()
new_info_df
```

```python
new_info_df.rename(columns = {'Series_Title':'title','IMDB_Rating':'imdb_rating','Meta_score':'meta_score','Director':'director','Star1':"star1",'Star2':'star2','No_of_Votes':'no_of_votes'}, inplace = True)
new_info_df.head()
```

### Connect to local database

```python
protocol = 'postgresql'
username = 'postgres'
password = 'postgres'
host = 'localhost'
port = 5432
database_name = 'movie_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

### Check for tables

```python
engine.table_names()
```

### Use pandas to load csv converted DataFrame into database

```python
new_money_df.to_sql(name='movie_money', con=engine, if_exists='append', index=False)
```

```python
new_info_df.to_sql(name='movie_info', con=engine, if_exists='append', index=False)
```

### Confirm data has been added by querying the customer_name table
* NOTE: can also check using pgAdmin

```python
pd.read_sql_query('select * from movie_money', con=engine).head()
```

```python
pd.read_sql_query('select * from movie_info', con=engine).head()
```
