```python
import pandas as pd
```

```python
df = pd.read_csv("cleaned_movie.csv")
```

```python
df
```

```python
#make bins
bins = [["20th Century", "20th Century Fox-Blue Sky Studios-20th Century Fox Animation"], 
        ["Marvel Studios"], 
        ["Walt Disney Pictures-Pixar", "Walt Disney Pictures", "Walt Disney Pictures-Walt Disney Animation Studios", "Walt Disney Pictures-Walt Disney Feature Animation", "Walt Disney Productions"], 
        ["Imagine Entertainment-Universal Pictures", "Universal Pictures", "Universal Pictures-Amblin Entertainment"], 
       ["Malpaso Productions-Warner Bros. Pictures", "Warner Bros. Pictures", "Warner Bros. Pictures-Morgan Creek Productions"], ["Paramount"], ["DreamWorks Animation"], ["Metro-Goldwyn-Mayer"]]

names = ["20th Century Fox", "Marvel Studios", "Disney", "Universal Pictures", "Warner Bros. Pictures", "Paramount", "DreamWorks Animation", "MGM"]
```

```python
studio_final = (pd.DataFrame({'Test': names, 'production_companies': bins})
        .explode('production_companies')
        .reset_index(drop=True))
```

```python
studios = df.merge(studio_final, on ="production_companies", how = "left").fillna("other")
studios.head(10)
```

```python
studios.to_csv('data.csv')
```
